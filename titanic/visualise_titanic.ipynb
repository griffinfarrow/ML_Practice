```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

This is a notebook to do some exploratory data visualisation on the Titanic Kaggle dataset to get a bit of a handle on what is going on 

```python
x = pd.read_csv("train.csv")
```

```python
x.head()
```

```python
# basic histograms showing our variables

fig,ax = plt.subplots(2, 2, figsize = (12, 6))
sns.histplot(x['Age'], ax=ax[0,0])
sns.histplot(x['Pclass'], ax=ax[0,1])
sns.histplot(x['Survived'], ax=ax[1,0])
sns.histplot(x['Sex'], ax=ax[1,1])
fig.tight_layout()
```

```python
# correlations between all variables
dataplot = sns.heatmap(x.corr(), cmap="plasma", annot=True)
```

Displays correlations between all variables
* Strong negative correlation between Fare and Pclass - lower fare means higher class (class 3 is worse than class 1)
* Strong positive correlation between Fare and Survived - is Fare a strong factor for deciding whether people survive?
* Strong negative correlation between Survived and Pclass - class strong factor in determining whether people survive?
